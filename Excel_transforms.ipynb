```python
import pandas as pd

# Загружаем файл
file_path = "/content/drive/MyDrive/Datasets_diplom/real_data_high.xlsx"
df = pd.read_excel(file_path, sheet_name="real_data_high")

# Группируем и разворачиваем строки в новые столбцы
df_grouped = df.groupby("Label").apply(lambda x: x.values.flatten())
df_transformed = pd.DataFrame(df_grouped.tolist())

# Создаем новые заголовки
new_columns = []
for i in range(3):  # Если у каждой группы по 3 строки
    new_columns.extend([f"{col}_{i+1}" for col in df.columns])

df_transformed.columns = new_columns

# Сохраняем
df_transformed.to_excel("transformed_data.xlsx", index=False)
```
